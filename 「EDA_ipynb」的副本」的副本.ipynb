```
data %>%
count(Country) %>%
ggplot(aes(x=Country,y=n,fill=Country))+geom_col()+geom_text(aes(label=n),vjust=-0.25)+labs(title="不同國家獲得世界首富前十對比",x="國家",y="數量")+theme(legend.position = "none")
```

```
data %>%
count(Country) %>%
ggplot(aes(x=fct_reorder(Country,n),y=n,fill=Country))+geom_col()+geom_text(aes(label=n),vjust=-0.25)+labs(title="不同國家獲得世界首富前十對比",x="國家",y="數量")+theme(legend.position = "none")
```
