```
install.packages("ggplot2")
library(ggplot2)
```

```
install.packages("SportsAnalytics")
library(SportsAnalytics)
```

```
NBA1617<-fetch_NBAPlayerStatistics("16-17")
```

```
str(NBA1617)
```

```
#利用order()函數,選一種類型的資料出來,由大到小排序(decreasing = T)
NBA1617Order<-NBA1617[order(NBA1617$ThreesMade,decreasing = T),]
NBA1617Order[1:20,]
```

```
#用subset()做欄位條件搜尋
subset(NBA1617,Team=="HOU")
```

```
#用grepl()做字串條件搜尋
NBA1617[grepl("Kevin",NBA1617$Name),]
```

```
#用ggplot畫出,四種不同類型的圖表(Exp:散佈圖、長條圖、折線圖...)
ggplot(NBA1617,aes(x=TotalPoints))+geom_histogram(bins=30,aes(color=Team))  
# X座標:總分 Y座標:資料件數 顏色:依不同球隊名而區別
```

```
ggplot(NBA1617,aes(x=Team))+geom_bar() # X座標:各隊伍的名稱
```

```
ggplot(NBA1617,aes(x=Position,y=ThreesMade))+geom_boxplot() # X座標:不同的守備位置 Y座標:三分球得分
```

```
ggplot(NBA1617,aes(x=ThreesAttempted,y=ThreesMade))+geom_point(aes(shape=Position,colour=Position))
# X座標:三分球進攻次數 Y座標:三分球得分 顏色:依照不同守備位置而區分
```
