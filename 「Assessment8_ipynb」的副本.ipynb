```
new_words <- c("慢性炎症","南方醫科大學","南方醫院","沉默殺手","中國營養與健康調查（CHNS）隊列研究中","第一四分位數","奇點糕","130/80mmHg","三高")
for(i in 1:length(new_words)){new_user_word(cutter,new_words[i])}

content <- str_remove_all(content,"[0-9a-zA-Z]+?")
cutter[content]
```

```
stop_words <- c("與","的","吃","了","是","不是","等","所以","咯","咳","為","對","中","才行","而","也","就","較","余名","出","就","都","用","萬","多","在","說","就是","全","很","那","一個","上","但是","有","沒有","可能","高","得")
writeLines(stop_words,"stop_words.txt")
```

```
cutter <- worker(user = "new_words.txt",stop_word = "stop_words.txt",bylines = FALSE)
seg_words <- cutter[content]
seg_words
```

```
txt_freq <- freq(seg_words)
txt_freq <- arrange(txt_freq,desc(freq))
head(txt_freq)
```
